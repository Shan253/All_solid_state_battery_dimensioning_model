### calculate battery performance

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.legend_handler import HandlerTuple
```

```python
# Sample data for demonstration purposes
battery_types = ['ASSB_NMC811', 'ASSB_LFP', 'LIB_NMC811', 'LIB_LFP']
performance_metrics = {
    'Specific Energy (Wh/kg)': {
        'Baseline': [ASSB_NMC811_composition_baseline['Specific_energy'], ASSB_LFP_composition_baseline['Specific_energy'], 
                     LIB_NMC811_composition_baseline['Specific_energy'], LIB_LFP_composition_baseline['Specific_energy']],
        'Optimal': [ASSB_NMC811_composition_optimal['Specific_energy'], ASSB_LFP_composition_optimal['Specific_energy'], 
                    LIB_NMC811_composition_optimal['Specific_energy'], LIB_LFP_composition_optimal['Specific_energy']]
    },
    'Energy Density (Wh/L)': {
        'Baseline': [ASSB_NMC811_composition_baseline['Energy_density'], ASSB_LFP_composition_baseline['Energy_density'], 
                     LIB_NMC811_composition_baseline['Energy_density'], LIB_LFP_composition_baseline['Energy_density']],
        'Optimal': [ASSB_NMC811_composition_optimal['Energy_density'], ASSB_LFP_composition_optimal['Energy_density'], 
                    LIB_NMC811_composition_optimal['Energy_density'], LIB_LFP_composition_optimal['Energy_density']]
    }
}


# Marker types for each battery type
markers = ['o', 's', '^', 'd']  # Circle, square, triangle, diamond

# Create figure and axis
fig, ax = plt.subplots(figsize=(4, 3.5), dpi=1000)

# Loop over battery types to plot baseline and optimal data
baseline_handles = []
optimal_handles = []
for i, battery in enumerate(battery_types):
    # Baseline performance in light blue
    baseline_scatter = ax.scatter(performance_metrics['Specific Energy (Wh/kg)']['Baseline'][i], performance_metrics['Energy Density (Wh/L)']['Baseline'][i],
                                  color='#e7d4e8', s=100, label=' Baseline', marker=markers[i])
    # Optimal performance in darker blue
    optimal_scatter = ax.scatter(performance_metrics['Specific Energy (Wh/kg)']['Optimal'][i], performance_metrics['Energy Density (Wh/L)']['Optimal'][i],
                                 color='#1b7837', s=100, label='Optimal', marker=markers[i])
    baseline_handles.append(baseline_scatter)
    optimal_handles.append(optimal_scatter)
    # Draw arrows showing improvement
    ax.annotate('', xy=(performance_metrics['Specific Energy (Wh/kg)']['Optimal'][i], performance_metrics['Energy Density (Wh/L)']['Optimal'][i]),
                xytext=(performance_metrics['Specific Energy (Wh/kg)']['Baseline'][i], performance_metrics['Energy Density (Wh/L)']['Baseline'][i]),
                arrowprops=dict(arrowstyle="->", linestyle='dashed', color='gray'))

# Adding labels and title
ax.set_xlabel('Specific Energy (Wh/kg)', fontsize=12)
ax.set_ylabel('Energy Density (Wh/L)', fontsize=12)
# ax.set_title('Comparison of Battery Performance', fontsize=12)


plt.show()

# system_color = ['navy', '#9970ab', '#c2a5cf', '#e7d4e8', '#d9f0d3', '#a6dba0', '#5aae61', '#1b7837']
```
